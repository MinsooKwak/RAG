```python
import camelot
import fitz  # PyMuPDF
from PIL import Image
import os

# PDF 파일 경로 설정
pdf_file_path = "2306.05685v4.pdf"
save_dir = "extracted_tables_images"
os.makedirs(save_dir, exist_ok=True)

# 1. camelot으로 PDF에서 표 추출 및 좌표 정보 얻기
try:
    tables = camelot.read_pdf(pdf_file_path, pages='all', flavor='stream')  # 격자x (논문)
    print(f"총 감지된 테이블 수: {len(tables)}")
    if len(tables) == 0:
        print("테이블이 감지되지 않았습니다. PDF 파일이 올바른지, 테이블이 있는지 확인해 주세요.")
except Exception as e:
    print(f"테이블 추출 중 오류 발생: {e}")

# 2. PDF 파일 열기
try:
    document = fitz.open(pdf_file_path)
    print("PDF 파일이 성공적으로 열렸습니다.")
except Exception as e:
    print(f"PDF 파일 열기 중 오류 발생: {e}")

# 3. 추출된 각 테이블의 좌표를 사용해 이미지로 저장
for i, table in enumerate(tables):
    try:
        # 테이블의 좌표 정보 (left, top, right, bottom)
        left, top, right, bottom = table._bbox
        print(f"테이블 {i + 1} 좌표: left={left}, top={top}, right={right}, bottom={bottom}")

        # 좌표가 유효한지 확인 (0보다 작거나 None일 경우 기본값 설정)
        if left is None or top is None or right is None or bottom is None:
            print(f"테이블 {i + 1}의 좌표 정보가 유효하지 않습니다. 기본값으로 설정합니다.")
            left, top, right, bottom = 0, 0, 100, 100  # 기본값 설정

        # PyMuPDF의 Rect 객체로 변환 (y 좌표를 반대로 설정하여 조정)
        rect = fitz.Rect(left, top, right, bottom)

        # 페이지 크기 확인
        page_number = table.page  # 해당 테이블이 위치한 페이지
        page = document.load_page(page_number - 1)  # 페이지 번호는 0부터 시작하므로 -1
        page_width, page_height = page.rect.width, page.rect.height
        print(f"페이지 {page_number} 크기: {page_width}x{page_height}")

        # 좌표가 페이지 경계를 벗어나지 않도록 정규화
        if rect.x0 < 0: rect.x0 = 0
        if rect.y0 < 0: rect.y0 = 0
        if rect.x1 > page_width: rect.x1 = page_width
        if rect.y1 > page_height: rect.y1 = page_height
        print(f"정규화된 좌표: {rect}")

        # 지정된 클립 영역에 콘텐츠가 있는지 확인
        text_in_rect = page.get_text("text", clip=rect).strip()
        if not text_in_rect:
            print(f"테이블 {i + 1}의 영역에 텍스트 또는 콘텐츠가 없습니다. 건너뜁니다.")
            continue

        # 해당 좌표를 클립하여 이미지로 변환
        pix = page.get_pixmap(clip=rect, dpi=300)
        print(f"픽스맵 생성 성공. 픽스맵 크기: {pix.width}x{pix.height}")

        # 픽스맵이 올바르게 생성되었는지 확인
        if pix.width <= 0 or pix.height <= 0:
            print(f"테이블 {i + 1}: 비어있는 픽스맵이 생성되었습니다.")
            continue

        # 이미지 객체 생성
        image = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        print("이미지 객체 생성 성공.")

        # 이미지 파일명 및 경로 설정
        image_path = os.path.join(save_dir, f"table_{i+1}_page{page_number}.png")

        # 파일 객체 대신 파일 경로로 이미지 저장
        image.save(image_path)
        print(f"테이블 {i+1}이 이미지로 저장되었습니다: {image_path}\n")
    except Exception as e:
        print(f"테이블 {i+1}의 이미지 저장 중 오류 발생: {e}")

# PDF 파일 닫기
document.close()
```

```python
import camelot
import fitz  # PyMuPDF
from PIL import Image
import os

# PDF 파일 경로 설정
pdf_file_path = "2306.05685v4.pdf"
save_dir = "filtered_tables_images2"
os.makedirs(save_dir, exist_ok=True)

# 필터링 기준 설정
MIN_WIDTH = 50
MIN_HEIGHT = 50
MIN_ROWS = 2
MIN_COLS = 2
KEYWORDS = ["figure", "table", "data", "chart", "graph"]

# 좌표 확장을 위한 여백 설정
MARGIN = 10  # 좌표를 확장할 여백 (단위: pt)

# 1. camelot으로 PDF에서 테이블 추출 및 좌표 정보 얻기
try:
    tables = camelot.read_pdf(pdf_file_path, pages='all', flavor='stream')  # 격자x (논문)
    print(f"총 감지된 테이블 수: {len(tables)}")
    if len(tables) == 0:
        print("테이블이 감지되지 않았습니다. PDF 파일이 올바른지, 테이블이 있는지 확인해 주세요.")
except Exception as e:
    print(f"테이블 추출 중 오류 발생: {e}")

# 2. 테이블 필터링 함수 정의
def is_valid_table(left, top, right, bottom, min_width=MIN_WIDTH, min_height=MIN_HEIGHT):
    """테이블의 좌표와 크기를 기준으로 유효성을 판단합니다."""
    width = right - left
    height = bottom - top
    return width >= min_width and height >= min_height

def extract_keywords_from_table(table, keywords=KEYWORDS):
    """테이블 내에 포함된 키워드를 추출하여 문자열로 반환합니다."""
    table_text = ' '.join(table.df.astype(str).values.flatten())  # 모든 셀의 텍스트를 하나의 문자열로 결합
    found_keywords = [keyword for keyword in keywords if keyword.lower() in table_text.lower()]
    return "_".join(found_keywords) if found_keywords else "no_keyword"

def is_valid_cell_count(table, min_rows=MIN_ROWS, min_cols=MIN_COLS):
    """테이블의 행과 열 개수를 기준으로 유효성을 판단합니다."""
    rows, cols = table.df.shape
    return rows >= min_rows and cols >= min_cols

# 3. 필터링된 테이블 저장
filtered_tables = []

for i, table in enumerate(tables):
    left, top, right, bottom = table._bbox
    # 필터링 조건 확인
    if (is_valid_table(left, top, right, bottom) and is_valid_cell_count(table)):
        keywords_in_table = extract_keywords_from_table(table)
        if keywords_in_table != "no_keyword":
            filtered_tables.append((table, keywords_in_table))
            print(f"테이블 {i + 1}이 필터를 통과했습니다. 키워드: {keywords_in_table}")
        else:
            print(f"무시된 테이블 {i + 1}: 키워드가 포함되지 않음.")
    else:
        print(f"무시된 테이블 {i + 1}: 좌표({left}, {top}, {right}, {bottom}) | 크기: ({right-left}x{bottom-top}) | 셀: {table.df.shape}")

# 4. PDF 파일 열기
try:
    document = fitz.open(pdf_file_path)
    print("PDF 파일이 성공적으로 열렸습니다.")
except Exception as e:
    print(f"PDF 파일 열기 중 오류 발생: {e}")

# 5. 필터링된 각 테이블의 좌표를 사용해 이미지로 저장
for i, (table, keywords) in enumerate(filtered_tables):
    try:
        # 테이블의 좌표 정보 (left, top, right, bottom)
        left, top, right, bottom = table._bbox
        print(f"테이블 {i + 1} 좌표: left={left}, top={top}, right={right}, bottom={bottom}")

        # 좌표가 유효한지 확인 (0보다 작거나 None일 경우 기본값 설정)
        if left is None or top is None or right is None or bottom is None:
            print(f"테이블 {i + 1}의 좌표 정보가 유효하지 않습니다. 기본값으로 설정합니다.")
            left, top, right, bottom = 0, 0, 100, 100  # 기본값 설정

        # 좌표에 여백 추가
        left -= MARGIN
        top += MARGIN
        right += MARGIN
        bottom -= MARGIN

        # PyMuPDF의 Rect 객체로 변환 (y 좌표를 반대로 설정하여 조정)
        rect = fitz.Rect(left, top, right, bottom)

        # 페이지 크기 확인
        page_number = table.page  # 해당 테이블이 위치한 페이지
        page = document.load_page(page_number - 1)  # 페이지 번호는 0부터 시작하므로 -1
        page_width, page_height = page.rect.width, page.rect.height
        print(f"페이지 {page_number} 크기: {page_width}x{page_height}")

        # 좌표가 페이지 경계를 벗어나지 않도록 정규화
        if rect.x0 < 0: rect.x0 = 0
        if rect.y0 < 0: rect.y0 = 0
        if rect.x1 > page_width: rect.x1 = page_width
        if rect.y1 > page_height: rect.y1 = page_height
        print(f"정규화된 좌표: {rect}")

        # 지정된 클립 영역에 콘텐츠가 있는지 확인
        text_in_rect = page.get_text("text", clip=rect).strip()
        if not text_in_rect:
            print(f"테이블 {i + 1}의 영역에 텍스트 또는 콘텐츠가 없습니다. 건너뜁니다.")
            continue

        # 해당 좌표를 클립하여 이미지로 변환
        pix = page.get_pixmap(clip=rect, dpi=300)
        print(f"픽스맵 생성 성공. 픽스맵 크기: {pix.width}x{pix.height}")

        # 픽스맵이 올바르게 생성되었는지 확인
        if pix.width <= 0 or pix.height <= 0:
            print(f"테이블 {i + 1}: 비어있는 픽스맵이 생성되었습니다.")
            continue

        # 이미지 객체 생성
        image = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        print("이미지 객체 생성 성공.")

        # 이미지 파일명 및 경로 설정
        image_path = os.path.join(save_dir, f"table_{i+1}_page{page_number}_{keywords}.png")

        # 파일 객체 대신 파일 경로로 이미지 저장
        image.save(image_path)
        print(f"테이블 {i+1}이 이미지로 저장되었습니다: {image_path}\n")
    except Exception as e:
        print(f"테이블 {i + 1}의 이미지 저장 중 오류 발생: {e}")

# PDF 파일 닫기
document.close()
```

```python
import camelot
import fitz  # PyMuPDF
from PIL import Image
import os

# PDF 파일 경로 설정
pdf_file_path = "2306.05685v4.pdf"
save_dir = "filtered_tables_images3"
os.makedirs(save_dir, exist_ok=True)

# 필터링 기준 설정
MIN_WIDTH = 50
MIN_HEIGHT = 50
MIN_ROWS = 2
MIN_COLS = 2
EXPAND_MARGIN = 20  # 테이블 좌표를 확장할 여백 (단위: pt)

# 1. camelot으로 PDF에서 테이블 추출 (stream 모드 사용)
def extract_tables_from_pdf(pdf_path, row_tol=10, flag_size=True):
    """PDF에서 stream 모드를 사용하여 테이블을 추출합니다."""
    tables = []
    try:
        tables += camelot.read_pdf(
            pdf_path,
            pages='all',
            flavor='stream',
            row_tol=row_tol,
            flag_size=flag_size,
            strip_text='\n'
        )
        print(f"stream 모드로 감지된 테이블 수: {len(tables)}")
    except Exception as e:
        print(f"stream 모드 테이블 추출 중 오류 발생: {e}")

    return tables

# 2. 테이블 필터링 및 좌표 확장 함수 정의
def is_valid_table(table, min_width=MIN_WIDTH, min_height=MIN_HEIGHT, min_rows=MIN_ROWS, min_cols=MIN_COLS):
    """테이블의 크기와 셀 수를 기준으로 유효성을 판단합니다."""
    left, top, right, bottom = table._bbox
    width = right - left
    height = bottom - top
    rows, cols = table.df.shape
    return width >= min_width and height >= min_height and rows >= min_rows and cols >= min_cols

def expand_table_coordinates(table, page_width, page_height, margin=EXPAND_MARGIN):
    """테이블 좌표를 확장하여 잘린 부분을 방지합니다."""
    left, top, right, bottom = table._bbox
    left = max(0, left - margin)
    top = max(0, top - margin)
    right = min(page_width, right + margin)
    bottom = min(page_height, bottom + margin)
    return left, top, right, bottom

# 3. 추출된 테이블 필터링 및 좌표 확장
def filter_and_expand_tables(tables, pdf_path):
    """유효한 테이블을 필터링하고, 좌표를 확장합니다."""
    valid_tables = []
    document = fitz.open(pdf_path)

    for table in tables:
        if is_valid_table(table):
            page_number = table.page
            page = document.load_page(page_number - 1)  # 페이지 번호는 0부터 시작하므로 -1
            page_width, page_height = page.rect.width, page.rect.height
            expanded_bbox = expand_table_coordinates(table, page_width, page_height)
            table._bbox = expanded_bbox  # 확장된 좌표로 갱신
            valid_tables.append((table, page_number))
        else:
            print(f"무시된 테이블: 좌표({table._bbox}), 셀: {table.df.shape}")

    document.close()
    return valid_tables

# 4. 테이블을 이미지로 저장
def save_tables_as_images(pdf_path, tables, save_dir):
    """추출된 테이블을 이미지로 저장합니다."""
    document = fitz.open(pdf_path)

    for i, (table, page_number) in enumerate(tables):
        try:
            left, top, right, bottom = table._bbox
            print(f"테이블 {i + 1} 확장된 좌표: left={left}, top={top}, right={right}, bottom={bottom}")

            # PyMuPDF의 Rect 객체로 변환
            rect = fitz.Rect(left, top, right, bottom)

            # 페이지 로드
            page = document.load_page(page_number - 1)  # 페이지 번호는 0부터 시작하므로 -1
            page_width, page_height = page.rect.width, page.rect.height
            print(f"페이지 {page_number} 크기: {page_width}x{page_height}")

            # 좌표가 페이지 경계를 벗어나지 않도록 정규화
            if rect.x0 < 0: rect.x0 = 0
            if rect.y0 < 0: rect.y0 = 0
            if rect.x1 > page_width: rect.x1 = page_width
            if rect.y1 > page_height: rect.y1 = page_height
            print(f"정규화된 좌표: {rect}")

            # 지정된 클립 영역에 콘텐츠가 있는지 확인
            text_in_rect = page.get_text("text", clip=rect).strip()
            if not text_in_rect:
                print(f"테이블 {i + 1}의 영역에 텍스트 또는 콘텐츠가 없습니다. 건너뜁니다.")
                continue

            # 해당 좌표를 클립하여 이미지로 변환
            pix = page.get_pixmap(clip=rect, dpi=300)
            print(f"픽스맵 생성 성공. 픽스맵 크기: {pix.width}x{pix.height}")

            # 픽스맵이 올바르게 생성되었는지 확인
            if pix.width <= 0 or pix.height <= 0:
                print(f"테이블 {i + 1}: 비어있는 픽스맵이 생성되었습니다.")
                continue

            # 이미지 객체 생성
            image = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
            print("이미지 객체 생성 성공.")

            # 이미지 파일명 및 경로 설정
            image_path = os.path.join(save_dir, f"table_{i+1}_page{page_number}.png")

            # 파일 객체 대신 파일 경로로 이미지 저장
            image.save(image_path)
            print(f"테이블 {i + 1}이 이미지로 저장되었습니다: {image_path}")
        except Exception as e:
            print(f"테이블 {i + 1}의 이미지 저장 중 오류 발생: {e}")

    document.close()

# 5. 테이블 추출 및 필터링 후 이미지로 저장
tables = extract_tables_from_pdf(pdf_file_path)
expanded_tables = filter_and_expand_tables(tables, pdf_file_path)
save_tables_as_images(pdf_file_path, expanded_tables, save_dir)
```
